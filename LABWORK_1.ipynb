### 1. Формулировка бизнес-задачи

**Бизнес-задача:**  
Выявить закономерности между аудио-характеристиками трека (темп, энергичность, танцевальность, тональность) и его популярностью среди слушателей. На основе исторических данных Spotify определить ключевые факторы, способствующие попаданию трека в чарты.

**Цель:**  
Подготовить данные для построения модели, способной предсказывать метрику `popularity` (от 0 до 100) на основе объективных параметров трека.

### 2. Гипотезы о факторах, влияющих на решение задачи

#### Гипотеза №1: "Формула танцевального хита"
Существует положительная корреляция между танцевальностью, энергичностью и популярностью, при этом затянутые треки теряют своего слушателя.
*   **Утверждение:** Треки с высокими показателями `danceability` (> 0.6) и `energy` (> 0.6), но с длительностью `duration_ms` ниже средней, имеют более высокий рейтинг `popularity`.

#### Гипотеза №2: "Влияние жанра и настроения (лада)"
Популярность зависит не только от качества звука, но и от жанра и эмоциональной окраски (мажор или минор).
*   **Утверждение:** Определенные жанры, например `pop`, в мажорной тональности статистически популярнее, чем нишевые жанры в миноре.

### 3. Описание набора данных (Словарь признаков)

В работе используется датасет **Spotify Tracks Dataset**. Ниже приведено описание каждого признака.

*   **`track_id`**: Уникальный ID трека в Spotify.
*   **`artists`**: Имена исполнителей (если их несколько, разделены символом ';').
*   **`album_name`**: Название альбома.
*   **`track_name`**: Название трека.
*   **`popularity`**: Популярность трека (0-100). Рассчитывается алгоритмом на основе количества и давности прослушиваний.
*   **`duration_ms`**: Продолжительность трека в миллисекундах.
*   **`explicit`**: Наличие ненормативной лексики (True/False).
*   **`danceability`**: Танцевальность (0.0-1.0). Насколько трек подходит для танцев.
*   **`energy`**: Энергичность (0.0-1.0). Воспринимаемая интенсивность и активность трека.
*   **`key`**: Тональность трека (0=C, 1=C#, 2=D, ...).
*   **`loudness`**: Общая громкость трека в децибелах (дБ).
*   **`mode`**: Настроение (1=Мажор, 0=Минор).
*   **`speechiness`**: Речеподобность (0.0-1.0). Наличие в треке произносимых слов (рэп, аудиокнига).
*   **`acousticness`**: Акустичность (0.0-1.0). Уверенность в том, что трек является акустическим.
*   **`instrumentalness`**: Инструментальность (0.0-1.0). Вероятность отсутствия вокала в треке.
*   **`liveness`**: Концертность (0.0-1.0). Вероятность того, что трек записан вживую с аудиторией.
*   **`valence`**: Позитивность (0.0-1.0). Музыкальная позитивность (счастье, эйфория).
*   **`tempo`**: Темп трека в ударах в минуту (BPM).
*   **`time_signature`**: Музыкальный размер (количество долей в такте). От 3 до 7 (3/4, 4/4, 5/4, 6/4, 7/4).
*   **`track_genre`**: Жанр трека.

---

### 4. Определение шкал измерения признаков

#### **1. Номинальная шкала**
*   **`track_id`**: Уникальный идентификатор записи.
*   **`artists`, `album_name`, `track_name`**: Текстовые данные (имена, названия).
*   **`track_genre`**: Жанры треков. Категории, которые нельзя упорядочить математически.
*   **`explicit`**: Бинарный признак (наличие или отсутствие цензуры).
*   **`mode`**: Категория лада (Мажор/Минор). 

#### **2. Порядковая шкала (Ordinal)**
*   **`key`**: Тональность (0 = C, 1 = C#, 2 = D...). Признак имеет четкий порядок, соответствующий возрастанию высоты звука в хроматической гамме, однако математические операции (сложение, деление) между тональностями не имеют физического смысла.

#### **3. Интервальная шкала**
*   **`loudness`**: Громкость в децибелах (дБ). Значение 0 дБ является условным порогом, а не отсутствием звука.
*   **`time_signature`**: Музыкальный размер (количество долей в такте). В музыкальной теории не существует значения "0 долей", шкала начинается с определенных значений (в датасете от 3 до 7). Мы можем измерить интервал (разница в 1 долю), но не можем говорить об абсолютном нуле.

#### **4. Относительная шкала**
*   **`popularity`**: Рейтинг (0-100). Значение 0 означает полное отсутствие популярности.
*   **`duration_ms`**: Физическая длительность. Существует абсолютный ноль (0 мс = отсутствие звука).
*   **`tempo`**: Темп (BPM). 0 BPM означает отсутствие ритма (абсолютный покой).
*   **`danceability`, `energy`, `speechiness`, `acousticness`, `instrumentalness`, `liveness`, `valence`**: Коэффициенты (0.0–1.0). Ноль означает полное отсутствие измеряемого свойства (например, нулевая энергия).
