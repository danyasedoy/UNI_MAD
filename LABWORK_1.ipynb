### 2. Гипотезы 

#### Гипотеза №1: "Формула хита"
Существует положительная корреляция между танцевальностью, энергичностью и популярностью, при этом затянутые треки теряют своего слушателя.
*   **Пояснение:** Треки с высокими показателями `danceability` (> 0.6) и `energy` (> 0.6), но с длительностью `duration_ms` ниже средней, имеют более высокий рейтинг `popularity`.

#### Гипотеза №2: "Влияние жанра и настроения"
Популярность зависит не только от качества звука, но и от жанра и эмоциональной окраски (мажор или минор).
*   **Пояснение:** Определенные жанры, например `pop`, в мажорной тональности статистически популярнее, чем нишевые жанры в миноре.

### 3. Описание набора данных (Словарь признаков)

В работе используется датасет **Spotify Tracks Dataset**. Ниже приведено описание каждого признака.

*   **`track_id`**: Уникальный ID трека в Spotify.
*   **`artists`**: Имена исполнителей (если их несколько, разделены символом ';').
*   **`album_name`**: Название альбома.
*   **`track_name`**: Название трека.
*   **`popularity`**: Популярность трека (0-100). Рассчитывается алгоритмом на основе количества и давности прослушиваний.
*   **`duration_ms`**: Продолжительность трека в миллисекундах.
*   **`explicit`**: Наличие ненормативной лексики (True/False).
*   **`danceability`**: Танцевальность (0.0-1.0). Насколько трек подходит для танцев.
*   **`energy`**: Энергичность (0.0-1.0). Воспринимаемая интенсивность и активность трека.
*   **`key`**: Тональность трека (0=C, 1=C#, 2=D, ...).
*   **`loudness`**: Общая громкость трека в децибелах (дБ).
*   **`mode`**: Настроение (1=Мажор, 0=Минор).
*   **`speechiness`**: Речеподобность (0.0-1.0). Наличие в треке произносимых слов (рэп, аудиокнига).
*   **`acousticness`**: Акустичность (0.0-1.0). Уверенность в том, что трек является акустическим.
*   **`instrumentalness`**: Инструментальность (0.0-1.0). Вероятность отсутствия вокала в треке.
*   **`liveness`**: Концертность (0.0-1.0). Вероятность того, что трек записан вживую с аудиторией.
*   **`valence`**: Позитивность (0.0-1.0). Музыкальная позитивность (счастье, эйфория).
*   **`tempo`**: Темп трека в ударах в минуту (BPM).
*   **`time_signature`**: Музыкальный размер (количество долей в такте). От 3 до 7 (3/4, 4/4, 5/4, 6/4, 7/4).
*   **`track_genre`**: Жанр трека.

---

### 4. Определение шкал измерения признаков

#### **1. Номинальная шкала**
*   **`track_id`**: Уникальный идентификатор записи.
*   **`artists`, `album_name`, `track_name`**: Текстовые данные (имена, названия).
*   **`track_genre`**: Жанры треков. Категории, которые нельзя упорядочить математически.
*   **`explicit`**: Бинарный признак (наличие или отсутствие цензуры).
*   **`mode`**: Категория лада (Мажор/Минор). 

#### **2. Порядковая шкала (Ordinal)**
*   **`key`**: Тональность (0 = C, 1 = C#, 2 = D...). Признак имеет четкий порядок, соответствующий возрастанию высоты звука в хроматической гамме, однако математические операции (сложение, деление) между тональностями не имеют физического смысла.

#### **3. Интервальная шкала**
*   **`loudness`**: Громкость в децибелах (дБ). Значение 0 дБ является условным порогом, а не отсутствием звука.
*   **`time_signature`**: Музыкальный размер (количество долей в такте). В музыкальной теории не существует значения "0 долей", шкала начинается с определенных значений (в датасете от 3 до 7). Мы можем измерить интервал (разница в 1 долю), но не можем говорить об абсолютном нуле.

#### **4. Относительная шкала**
*   **`popularity`**: Рейтинг (0-100). Значение 0 означает полное отсутствие популярности.
*   **`duration_ms`**: Физическая длительность. Существует абсолютный ноль (0 мс = отсутствие звука).
*   **`tempo`**: Темп (BPM). 0 BPM означает отсутствие ритма (абсолютный покой).
*   **`danceability`, `energy`, `speechiness`, `acousticness`, `instrumentalness`, `liveness`, `valence`**: Коэффициенты (0.0–1.0). Ноль означает полное отсутствие измеряемого свойства (например, нулевая энергия).

```python

import pandas as pd

df = pd.read_csv('dataset.csv')

print("Данные успешно загружены! Вот первые 5 строк:")
df.head()
```

```python
print("Общая информация о датасете:")
df.info()
```

```python
# Удаляем 'Unnamed: 0'
if 'Unnamed: 0' in df.columns:
    df.drop('Unnamed: 0', axis=1, inplace=True)

# Удаляем строки с пропусками
df.dropna(inplace=True)

# Проверка на дубликаты
duplicates_count = df.duplicated().sum()
print(f"Количество полных дубликатов: {duplicates_count}")

# Если дубликаты есть, удалим их
if duplicates_count > 0:
    df.drop_duplicates(inplace=True)
    print("Дубликаты удалены.")

# Проверяем результат очистки
df.info()
```

### 4.5. Избавляемся от огромной размерности жанров
* Посмотрим на общий список жанров
* Разобъем их на 4 группы для One-Hot-Encoding кодирования

```python
unique_genres = df['track_genre'].unique()
print(f"Всего уникальных жанров: {len(unique_genres)}")
print(sorted(list(unique_genres)))
```

```python
pop_keywords = ['pop', 'dance', 'disco', 'electro', 'indie', 'acoustic', 'house', 'techno']
rock_keywords = ['rock', 'metal', 'punk', 'grunge', 'emo', 'goth', 'alternative', 'psych', 'hard', 'heavy', 'reggae']
rap_keywords = ['rap', 'trap', 'r-n-b', 'funk', 'hop', 'dub']

def classify_genre(genre):
    if any(keyword in genre for keyword in rap_keywords):
        return 'Rap'
    elif any(keyword in genre for keyword in rock_keywords):
        return 'Rock'
    elif any(keyword in genre for keyword in pop_keywords):
        return 'Pop'
    else:
        return 'Other'

df['macro_genre'] = df['track_genre'].apply(classify_genre)
```

### 5. Масштабирование признаков и кодирование категориальных признаков
#### Нормализация
*  **`duration_ms`**,**`tempo`**,**`loudness`**,**`danceability`**,**`energy`**,**`speechiness`**,**`acousticness`**,**`instrumentalness`**,**`liveness`**,**`valence`**,**`time_signature`**, приведем к диапазону [0,1]
*  **`popularity`** исключаем так как это целевая колонка
#### Кодирование категориальных признаков
* Кодируем **`explicit`** как категориальный признак (из True/False в 0/1)
* Проводим One-Hot Encoding для 'track_genre' И для 'key' 

```python
from sklearn.preprocessing import MinMaxScaler

# Выбираем числовые колонки для Нормализации 
# Исключаем: 'popularity' (целевая), 'key' и 'mode' (категории), 'explicit' (бинарный)
num_cols = [
    'duration_ms', 'tempo', 'loudness', 
    'danceability', 'energy', 'speechiness', 'acousticness', 
    'instrumentalness', 'liveness', 'valence', 'time_signature'
]

# Создаем копию для нормализации
df_processed = df.copy()

# Применяем MinMax Scaling
scaler = MinMaxScaler()
df_processed[num_cols] = scaler.fit_transform(df[num_cols])

# Кодирование категориальных признаков
df_processed['explicit'] = df_processed['explicit'].astype(int)

# One-Hot Encoding для 'macro_genre' И для 'key' 
df_encoded = pd.get_dummies(df_processed, columns=['macro_genre', 'key'], dtype=int)

# Удаляем столбец 'track-genre'
df_encoded = df_encoded.drop('track_genre', axis=1)

# Выводим результат нормализации
print("Пример масштабирования (Loudness):")
print(f"Было: {df['loudness'].min()} ... {df['loudness'].max()}")
print(f"Стало: {df_encoded['loudness'].min()} ... {df_encoded['loudness'].max()}")

# Выводим результат One-Hot Encoding
print("Пример кодирования:")
key_cols = [c for c in df_encoded.columns if c.startswith('macro_genre_')]
print(df_encoded[key_cols].head(3))
```

```python
df_encoded.to_csv('dataset_normalized.csv', index=False)

print("Файлы успешно сохранены!")
```

### 5.5 Стандартизация
* Применяем StandartScaler для того же набора данных
* Проводим кодирование категориальных признаков
* Сохраняем новый стандартизированный датасет

```python
from sklearn.preprocessing import StandardScaler

df_st = df.copy()

st_scaler = StandardScaler()
df_st[num_cols] = st_scaler.fit_transform(df[num_cols])

df_st['explicit'] = df_st['explicit'].astype(int)

df_st = pd.get_dummies(df_st, columns=['macro_genre', 'key'], dtype=int)

df_st = df_st.drop('track_genre', axis=1)

df_st.to_csv('dataset_standardized.csv', index=False)
print("Файлы успешно сохранены!")
```
